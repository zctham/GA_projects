# Cleaning of Reviews Dataset

1. Cleaning
    - Dropping nulls, rows only with special characters
    - Dropping alphanumeric values
    - Dropping host canceled listings
    - Remove html breaks
    - Dropping non english comments with langdetect library

```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

import time
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)

# pip install langdetect
from langdetect import detect, detect_langs, DetectorFactory

from nltk.sentiment.vader import SentimentIntensityAnalyzer

import warnings
warnings.filterwarnings("ignore")
```

```python
reviews = pd.read_csv('../assets/reviews.csv')
sum_reviews = pd.read_csv('../assets/sum_reviews.csv')
```

```python
reviews
```

```python
reviews.info()
```

```python
%%time
# Dropping null comments

reviews = reviews.dropna(subset = ['comments'], axis = 0)
reviews.shape

# Dropping comments with '-'

drop_dash = reviews[reviews['comments'].str.match('-')].index
reviews = reviews.drop(drop_dash)

# Drop host cancelled listings
canc_list = reviews[reviews['comments'].str.match('The host canceled this reservation')].index
reviews = reviews.drop(canc_list)

# Dropping comments with breaks 
# breaks = reviews[reviews['comments'].str.match('(<|&lt;)br\s*\/*(>|&gt;)')].index
# reviews = reviews.drop(breaks)

# for row in reviews['comments']:
#     row = re.sub('<.*/?>', r'', row)

# Dropping alphanumeric values 
reviews['comments'] = reviews['comments'].map(lambda x: re.sub('\w*\d\w*','',x))

# Reset index
reviews = reviews.reset_index(drop = True)

reviews
```

```python
%%time
# Cleaning html breaks
for number, value in enumerate(reviews['comments']):
    reviews['comments'][number] = re.sub(r'<.*?>', r'', value)
    
reviews['comments']
```

```python
%%time
num_errors = 0
err = []

for index, row in reviews.iterrows():
    try:
        language = detect(row['comments']) #(" ".join(row.split())) # row.strip())
    except:
        num_errors += 1
        err.append(index)
        print("This row throws an error:", row['comments'])
```

```python
reviews = reviews.drop(err)
reviews
```

```python
%%time
# Highlighting language of comments in new column and dropping non english columns
DetectorFactory.seed = 0
reviews['language'] = reviews['comments'].apply(detect)

reviews
```

```python
plt.figure(figsize = (10,8))
reviews['language'].value_counts(ascending = False).plot(kind = 'barh')
```

```python
# Filter out for english language, drop the other languages
reviews = reviews[reviews['language'] == 'en']

# Reset index
reviews = reviews.reset_index(drop = True)
```
