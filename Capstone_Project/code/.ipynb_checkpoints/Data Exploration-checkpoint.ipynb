### Cleaning comments column

```python
%%time
# Dropping null comments

reviews = reviews.dropna(subset = ['comments'], axis = 0)
reviews.shape

# Dropping comments with '-'

drop_dash = reviews[reviews['comments'].str.match('-')].index
reviews = reviews.drop(drop_dash)

# Drop host cancelled listings
canc_list = reviews[reviews['comments'].str.match('The host canceled this reservation')].index
reviews = reviews.drop(canc_list)

# Dropping comments with breaks 
# breaks = reviews[reviews['comments'].str.match('(<|&lt;)br\s*\/*(>|&gt;)')].index
# reviews = reviews.drop(breaks)

# for row in reviews['comments']:
#     row = re.sub('<.*/?>', r'', row)

# Dropping alphanumeric values 
reviews['comments'] = reviews['comments'].map(lambda x: re.sub('\w*\d\w*','',x))

# Reset index
reviews = reviews.reset_index(drop = True)

reviews
```

```python
%%time
for number, value in enumerate(reviews['comments']):
    reviews['comments'][number] = re.sub(r'<.*?>', r'', value)
    
reviews['comments']
```

### Dropping non english comments

```python
%%time
num_errors = 0
err = []

for index, row in reviews.iterrows():
    try:
        language = detect(row['comments']) #(" ".join(row.split())) # row.strip())
    except:
        num_errors += 1
        err.append(index)
        print("This row throws an error:", row['comments'])
```

```python
reviews = reviews.drop(err)
reviews
```

```python
%%time
# Highlighting language of comments in new column and dropping non english columns
DetectorFactory.seed = 0
reviews['language'] = reviews['comments'].apply(detect)

reviews
```

## NLTK's VADER for Sentiment Analysis

```python
%%time
# NLTK's VADER uses polarity (positive and negative) and intensity of emotions/words 

# Initial test of VADER
sent = SentimentIntensityAnalyzer()

reviews['scores'] = reviews['comments'].apply(lambda review: sent.polarity_scores(review))

reviews.head()
```

```python
# Getting compound score

reviews['compound']  = reviews['scores'].apply(lambda score_dict: score_dict['compound'])

reviews.head()
```

```python
reviews.describe()
```

```python
# Exploring negative and positive scores and comments

pos = {'comments':reviews['comments'],
       'compound':reviews['compound']
      }
       
pos_df = pd.DataFrame(pos).sort_values('compound',ascending = False)
pos_df.head()
```
