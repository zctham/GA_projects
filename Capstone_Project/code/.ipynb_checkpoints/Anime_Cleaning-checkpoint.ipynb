# Introduction

## Problem Statement

Recommender systems has been a huge portion of our lives. In the last few decades alone, the rise of Youtube, Amazon, Spotify, Netflix and other similar web services are unavoidable in our daily online journeys. The significance of recommender systems is further emphasised in the famous Netflix challenge in 2006, where top coders were given the challenge of improving Netflix's own recommender system algorithm, with the grand prize of 1 million dollars given to whichever team is able to get a higher RMSE score.

In a very general way, recommender systems are algorithms aimed at suggesting relevant items to users (items being movies to watch, text to read, products to buy or anything else depending on industries). When users feel that items recommended are relevant to them, there is higher propensity for users to 'like' or purchase an item, which in turn increases viewership or revenue for companies. It is not far-fetched to say that recommender systems play a huge part in ensuring a companies' survival and continuity.

In this capstone project, we will be looking into recommendation system for anime titles. While this may not translate into earnings of any sort for any company, the algorithm used will be largely similar to what is being deployed in industries today. Measure of success will be evaluated with a RMSE score and a metric 'MAP@K' or 'Mean Average Precision at k', which translates to how many anime titles are relevant out of 'k' number of anime titles recommended.

## Data Cleaning and EDA

We will be merging the anime dataset and rating dataset before doing our cleaning and EDA.

```python
# Merging rating and anime data

anime_comb = anime.merge(rating, on = 'anime_id')
anime_comb
```

```python
anime_comb.info()
```

### Anime data

#### Synopsis

```python
# Merging anime with synopsis
synopsis = synopsis.rename(columns = {'MAL_ID': 'anime_id', 'Name':'name'})

anime_comb = anime_comb.merge(synopsis, on = ['name'])
anime_comb = anime_comb.rename(columns = {'sypnopsis':'synopsis'})
anime_comb.head()
```

There are some synopsis that starts with 'No synopsis information..', and we will be dropping such titles as we are looking to use synopsis as one of the features for content filtering.

```python
# Drop rows with no synopsis information as synopsis is one of the features we will be using for content based recommender
no_info = anime_comb[anime_comb['synopsis'].str.match('No synopsis information')].index
anime_comb = anime_comb.drop(no_info)

anime_comb
```

```python
anime_comb.shape
```

About 700,000+ rows are dropped.

#### Genre

We will drop null values in 'genre' column.

```python
# Drop rows with null genres
anime_comb = anime_comb.dropna(subset = ['genre'])

anime_comb
```

To get the count of total genres by its type, we explode the genre column and plot the count of the type of genres.

```python
# Converting genres to list data
anime_comb['genre'] = anime_comb['genre'].str.split(',')

anime_comb
```

```python
%%time
# Exploring the number of animes by each genre by exploding dataset via 'Genre'
def remove_whitespace(x):
        return x.strip()
    
exploded = anime_comb.explode('genre', ignore_index = True)

exploded['genre'] = exploded['genre'].map(remove_whitespace)
```

```python
# Plotting top genres
exploded['genre'].value_counts(ascending = True).plot(kind='barh',figsize = (25,15))
plt.yticks(fontsize=15, rotation=0);
plt.title('Top Genres by count', fontsize = 30)
```

```python
all_genres = defaultdict(int)

for genres in anime_comb['genre']:
    for genre in genres:
        all_genres[genre.strip()] += 1
```

```python
genres_cloud = WordCloud(width=900, height=500, background_color='white', colormap='gnuplot').generate_from_frequencies(all_genres)
plt.figure(figsize = (10,8))
plt.imshow(genres_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Top genres are 'Comedy', 'Action', 'Romance', 'Fantasy' and 'Drama'.

```python
# Plotting top types of anime
types = anime_comb['type'].value_counts()
plt.figure(figsize = (10,8))

sns.barplot(x = types.values,
           y = types.index,
            data = anime,
            palette = 'viridis'
           )

plt.title('Top types by count', fontsize = 15)
```

'TV' has the most count under types and 'music' has the least. 'TV', 'OVA' and 'Special' are what most users usually watch, typically lasting 25 minutes. 

```python
# Plotting types of animes of their ratings

plt.figure(figsize = (10,8))
sns.boxplot(data = anime_comb, y = 'rating_x', x = 'type')
```

There is no clear conclusion on what types have a higher rating, but 'Movie' type appears to have a slightly better range of scores compared to other types.

#### Episodes

```python
# Drop rows with unknown episodes
anime_comb = anime_comb[anime_comb['episodes'] != 'Unknown']
```

```python
anime_comb['episodes'].value_counts()
```

#### Name

We will clean up the anime titles to remove special characters.

```python
# Cleanup name column
def text_cleaning(text):
    text = re.sub(r'&quot;', '', text)
    text = re.sub(r'.hack//', '', text)
    text = re.sub(r'&#039;', '', text)
    text = re.sub(r'A&#039;s', '', text)
    text = re.sub(r'I&#039;', 'I\'', text)
    text = re.sub(r'&amp;', 'and', text)
    
    return text

anime_comb['name'] = anime_comb['name'].apply(text_cleaning)

anime_comb
```

We will do some cleaning up of our anime titles to remove special characters.

```python
# Drop rows with unknown episodes and convert column to int
anime_comb = anime_comb[anime_comb['episodes'] != 'Unknown']
anime_comb['episodes'] = anime_comb['episodes'].astype('int64')

# Reset index
anime_comb = anime_comb.reset_index(drop = True)

anime_comb
```
