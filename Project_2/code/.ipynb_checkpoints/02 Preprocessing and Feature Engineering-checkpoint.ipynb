## Introduction

In this notebook, we will be executing what was covered in the previous notebook.

1. Create new columns and drop used columns
2. Imputing null values
3. Converting ordinal columns to numerical data
4. Converting numerical columns to categorical data
5. Dropping columns that have high single value counts
6. Remove outliers
7. Dropping columns that are collinear with each other

We will also be performing log transformation for saleprice variable and some other variables due to skewness.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_style('ticks')

pd.set_option('display.max_columns', None)

%matplotlib inline
```

```python
# Graph functions

# Function for scatterplots

def scatterplot(data, x, y):
    plt.scatter(data, x, y)

# Function for plotting boxplots

def subplot_box(dataframe, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns)/4)) 
    fig, ax = plt.subplots(nrows=nrows, ncols=4,figsize=(15, nrows*3.5)) 
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        sns.boxplot(x = dataframe[column], y = dataframe.saleprice, width = 0.3, ax = ax[i])
        #ax[i].set_title(column,fontsize=14)  
        ax[i].tick_params(labelsize=12)
        ax[i].set_xlabel(column, fontsize=14)
       #ax[i].set_ylabel('saleprice', fontsize=14)
    plt.tight_layout()
    
# Function for histogram subplots
def subplot_histograms(dataframe, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns)/4)) 
    fig, ax = plt.subplots(nrows=nrows, ncols=4,figsize=(15, nrows*3.5)) 
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        ax[i].hist(dataframe[column],bins=15)
        ax[i].set_title(f'{column} distribution',fontsize=14)
        ax[i].tick_params(labelsize=12)
        ax[i].set_xlabel(column, fontsize=14)
    plt.tight_layout()
      
# Function for scatterplots 
def subplot_scatter(dataframe, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns)/4)) 
    fig, ax = plt.subplots(nrows=nrows, ncols=4,figsize=(15, nrows*3.5)) 
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        sns.regplot(y=dataframe.saleprice, x=dataframe[column],ax=ax[i], \
                    scatter_kws={'s':15},
                    line_kws = {'color':'orange'})
        ax[i].set_title(f'{column} vs saleprice',fontsize=14)  
        ax[i].tick_params(labelsize=12)
        ax[i].set_xlabel(column, fontsize=14)
        ax[i].set_ylabel('saleprice', fontsize=14)
    plt.tight_layout()

    
# Imputation functions 

# Impute with 'NA'

def impute_na(data, columns):
    for col in columns:
        data[col] = data[col].fillna('NA')

#Impute with 'None'

def impute_none(data,columns):
    for col in columns:
        data[col] = data[col].fillna('None')

# Impute with 0
        
def impute_zero(data,columns):
    for col in columns:
        data[col] = data[col].fillna(float(0))
        
# Function to drop columns

def drop_cols(data, cols_list):
    data.drop(cols_list, axis = 1, inplace = True)
```

```python
# Bringing in train and test data

train = pd.read_csv('../datasets/train.csv')
test = pd.read_csv('../datasets/test.csv')

train.columns = [col.lower().replace(' ','_') for col in train.columns]
test.columns = [col.lower().replace(' ','_') for col in test.columns]


display(train.head())
display(test.head())
```

As we will be applying the same cleaning to both train and test data, we will create a function that incorporates all the cleaning and feature engineering.

```python
def cleaning_engineering(data):
    
    ########## Create columns ##########
    
    def drop_cols(data,col_list):
        data.drop(col_list, axis = 1, inplace = True)
        
    # Create house age column
    for index, val in enumerate(data['year_built']):
        if val == data.loc[index, 'yr_sold']:
            data.loc[index, 'house_age'] = 0
        else:
            data.loc[index, 'house_age'] = data.loc[index,'yr_sold'] - val
    
    # Create column for porch features combined
    porch_cols = ['3ssn_porch','enclosed_porch','open_porch_sf','screen_porch']
    data['with_porch'] = data[porch_cols].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    drop_cols(data, porch_cols)
    
    # Create column for if house is remodelled
    data['remod/add'] = data['year_remod/add'] - data['year_built']
    data['remod/add'] = data['remod/add'].apply(lambda x: 1 if x > 0 else 0)
    drop_cols(data,'year_remod/add')
    
    # Create column for big house and remove 1st_flr_sf, 2nd_flr_sf
    data['big_house'] = data['1st_flr_sf'] + data['2nd_flr_sf'] - data['gr_liv_area']
    data['big_house'] = data['big_house'].apply(lambda x: 1 if x < 0 else 0)
    
    # Create column for new exterior_1st and drop exterior_2nd
    exterior_materials = {'VinylSd':'Vinyl',
                          'MetalSd': 'Metal',
                          'HdBoard': 'Wood',
                          'Wd Sdng': 'Wood',
                          'Plywood': 'Wood',
                          'WdShing': 'Wood',
                          'CemntBd': 'Mix',
                          'Stucco': 'Mix',
                          'CBlock': 'Mix',
                          'AsphShn': 'Mix',
                          'ImStucc': 'Mix',
                          'BrkFace': 'Brick',
                          'BrkComm': 'Brick',
                          'AsbShng': 'Aesbestos',
                          'Stone': 'Stone'}
    data['exterior'] = data['exterior_1st'].map(exterior_materials)
    drop_cols(data, ['exterior_1st','exterior_2nd'])
    
    ########## Impute null values ##########
    
    # Impute 'NA' and 'None' 
    na_cols = ['pool_qc','misc_feature','alley','fence','fireplace_qu','garage_finish','garage_qual','garage_cond',
          'garage_type','bsmt_cond','bsmt_exposure','bsmt_qual','bsmtfin_type_1','bsmtfin_type_2']

    none_cols = ['mas_vnr_type']
    
    impute_na(data,na_cols)
    impute_none(data,none_cols)
    
    # Impute zero for lot_frontage,'mas_nvr_area','bsmt_full_bath','bsmtfin_sf_1','bsmt_unf_sf','garage_area'
    zeros = ['lot_frontage','mas_vnr_area','bsmt_full_bath','bsmtfin_sf_1','bsmt_unf_sf','garage_area']
    impute_zero(data,zeros)
    
    
    ########## Converting ordinal columns to numerical data ##########
    
    # Function for converting to numerical values
    def convert_num(colname, dictionary):
        data[colname] = data[colname].map(dictionary)
    
    # Convert all 'NA' to 'None' to capture changes while converting ordinal columns to numerical data
    data.replace('NA','None',inplace=True)
    
    # Defining dictionary values for renaming
    qual = {'Ex': 5,
            'Gd': 4,
            'TA': 3,
            'Fa': 2,
            'Po': 1,
            'None': 0}
    
    lotshape = {'Reg': 0,
                'IR1': 1,
                'IR2': 2,
                'IR3': 3}
    
    bsmtex = {'None': 0,
              'No': 1,
              'Mn': 2,
              'Av': 3,
              'Gd': 4}
    
    # Map numerical values
    convert_num('lot_shape', dictionary = lotshape)
    convert_num('bsmt_exposure', dictionary = bsmtex)
    convert_num('exter_qual', dictionary = qual)
    convert_num('bsmt_qual', dictionary = qual)
    convert_num('heating_qc', dictionary = qual)
    convert_num('kitchen_qual', dictionary = qual)
    convert_num('fireplace_qu', dictionary = qual)
    
    ########## Converting numerical to categorical data ##########
    data['ms_subclass'] = data['ms_subclass'].astype('object')
    
    ########## Dropping columns that have high single value counts and collinearity ##########
    
    # We will also drop pid as it is unlikely to impact saleprice as seen in the scatterplots. 
    to_drop = ['pid','low_qual_fin_sf','bsmt_half_bath','kitchen_abvgr','pool_area','misc_val','street',
                'alley','utilities','land_slope','condition_2','roof_matl','bsmt_cond','bsmtfin_type_2','heating',
                'central_air','electrical','functional','garage_qual','paved_drive','pool_qc','fence','misc_feature',
                'sale_type','garage_cars','totrms_abvgrd','garage_yr_blt','total_bsmt_sf','bsmtfin_sf_2']
    
    drop_cols(data,to_drop)
    
    ########## Removing outliers ##########
    data.drop(data[data['lot_frontage'] > 300].index, inplace = True)
    data.drop(data[data['lot_area'] > 100000].index, inplace = True)
    data.drop(data[data['gr_liv_area'] > 4500].index, inplace = True)
    
    # Convert all 'NA' to 'None' to capture changes while converting ordinal columns to numerical data
    # data.replace('NA','None',inplace=True)
    
    # To reset index
    data = data.reset_index(drop = True)
    
    return data
```

```python
# Function to clean and feature engineer test data, except for removing rows / outliers as required by Kaggle to have 
# full dataset for testing

def cleaning_engineering_test(data):
    
    ########## Create columns ##########
    
    def drop_cols(data,col_list):
        data.drop(col_list, axis = 1, inplace = True)
        
    # Create house age column
    for index, val in enumerate(data['year_built']):
        if val == data.loc[index, 'yr_sold']:
            data.loc[index, 'house_age'] = 0
        else:
            data.loc[index, 'house_age'] = data.loc[index,'yr_sold'] - val
    
    # Create column for porch features combined
    porch_cols = ['3ssn_porch','enclosed_porch','open_porch_sf','screen_porch']
    data['with_porch'] = data[porch_cols].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    drop_cols(data, porch_cols)
    
    # Create column for if house is remodelled
    data['remod/add'] = data['year_remod/add'] - data['year_built']
    data['remod/add'] = data['remod/add'].apply(lambda x: 1 if x > 0 else 0)
    drop_cols(data,'year_remod/add')
    
    # Create column for big house and remove 1st_flr_sf, 2nd_flr_sf
    data['big_house'] = data['1st_flr_sf'] + data['2nd_flr_sf'] - data['gr_liv_area']
    data['big_house'] = data['big_house'].apply(lambda x: 1 if x < 0 else 0)
    
    # Create column for new exterior_1st and drop exterior_2nd
    exterior_materials = {'VinylSd':'Vinyl',
                          'MetalSd': 'Metal',
                          'HdBoard': 'Wood',
                          'Wd Sdng': 'Wood',
                          'Plywood': 'Wood',
                          'WdShing': 'Wood',
                          'CemntBd': 'Mix',
                          'Stucco': 'Mix',
                          'CBlock': 'Mix',
                          'AsphShn': 'Mix',
                          'ImStucc': 'Mix',
                          'PreCast': 'Mix',
                          'BrkFace': 'Brick',
                          'BrkComm': 'Brick',
                          'AsbShng': 'Aesbestos',
                          'Stone': 'Stone'}
    data['exterior'] = data['exterior_1st'].map(exterior_materials)
    drop_cols(data, ['exterior_1st','exterior_2nd'])
    
    ########## Impute null values ##########
    
    # Impute 'NA' and 'None' 
    na_cols = ['pool_qc','misc_feature','alley','fence','fireplace_qu','garage_finish','garage_qual','garage_cond',
          'garage_type','bsmt_cond','bsmt_exposure','bsmt_qual','bsmtfin_type_1','bsmtfin_type_2']

    none_cols = ['mas_vnr_type']
    
    impute_na(data,na_cols)
    impute_none(data,none_cols)
    
    # Impute zero for lot_frontage,'mas_nvr_area','bsmt_full_bath','bsmtfin_sf_1','bsmt_unf_sf','garage_area'
    zeros = ['lot_frontage','mas_vnr_area','bsmt_full_bath','bsmtfin_sf_1','bsmt_unf_sf','garage_area']
    impute_zero(data,zeros)
    
    
    ########## Converting ordinal columns to numerical data ##########
    
    # Function for converting to numerical values
    def convert_num(colname, dictionary):
        data[colname] = data[colname].map(dictionary)
    
    # Convert all 'NA' to 'None' to capture changes while converting ordinal columns to numerical data
    data.replace('NA','None',inplace=True)
    
    # Defining dictionary values for renaming
    qual = {'Ex': 5,
            'Gd': 4,
            'TA': 3,
            'Fa': 2,
            'Po': 1,
            'None': 0}
    
    lotshape = {'Reg': 0,
                'IR1': 1,
                'IR2': 2,
                'IR3': 3}
    
    bsmtex = {'None': 0,
              'No': 1,
              'Mn': 2,
              'Av': 3,
              'Gd': 4}
    
    # Map numerical values
    convert_num('lot_shape', dictionary = lotshape)
    convert_num('bsmt_exposure', dictionary = bsmtex)
    convert_num('exter_qual', dictionary = qual)
    convert_num('bsmt_qual', dictionary = qual)
    convert_num('heating_qc', dictionary = qual)
    convert_num('kitchen_qual', dictionary = qual)
    convert_num('fireplace_qu', dictionary = qual)
    
    ########## Converting numerical to categorical data ##########
    data['ms_subclass'] = data['ms_subclass'].astype('object')
    
    ########## Dropping columns that have high single value counts and collinearity ##########
    
    # We will also drop pid as it is unlikely to impact saleprice as seen in the scatterplots. 
    to_drop = ['pid','low_qual_fin_sf','bsmt_half_bath','kitchen_abvgr','pool_area','misc_val','street',
                'alley','utilities','land_slope','condition_2','roof_matl','bsmt_cond','bsmtfin_type_2','heating',
                'central_air','electrical','functional','garage_qual','paved_drive','pool_qc','fence','misc_feature',
                'sale_type','garage_cars','totrms_abvgrd','garage_yr_blt','total_bsmt_sf','bsmtfin_sf_2']
    
    drop_cols(data,to_drop)
    
    # To reset index
    data = data.reset_index(drop = True)
    
    return data
```

### Log transformation

We will be performing log transformation for saleprice due to it being right skewed. We will also look at other numerical variables and log transform them if they have skewness of above 0.5 to have more 'normal' looking distributions.

```python
# Log transformation for saleprice
train_processed.loc[:,'saleprice'] = np.log(train_processed['saleprice'])
train_processed['saleprice']
```

We see that the saleprice has a more normal distribution after log transformation.

```python
plt.figure(figsize = (8,5))
plt.hist(train_processed['saleprice'])
```

Next, we will perform log transformation for variables that have skewness of above 0.5.

```python
num_cols = train_processed.select_dtypes(exclude = 'object').columns
num_cols
```

```python
# Log transformation for skewed numerical columns

num_cols = train_processed.select_dtypes(exclude = 'object').columns

# Drop columns that are ordinal as it is meaningless to log transform them 
cols_to_drop = ['year_built','lot_shape','overall_qual','overall_cond','exter_qual','bsmt_qual','bsmt_exposure',
               'heating_qc','bsmt_full_bath','full_bath','half_bath','bedroom_abvgr','kitchen_qual','fireplaces',
               'fireplace_qu','mo_sold','yr_sold','with_porch','remod/add','big_house']

num_cols = num_cols.drop(cols_to_drop)

skewed_features = train_processed[num_cols].skew()

skewed = skewed_features[abs(skewed_features) > 0.5]

print("{} numerical features with skew > 0.5 :".format(len(skewed)))
print(skewed)
```

```python
# Log transformation for skewed numerical columns

for ind in skewed.index:
    train_processed.loc[:,ind] = np.log1p(train_processed[ind])
```

We can see that the values have a more normal distribution after log transformation. Values of zero before transformation will remain zero as they cannot be transformed.
